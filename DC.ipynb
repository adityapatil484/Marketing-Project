```python
watched_movie = df[df['Have you seen this DC movie?'] == 'Yes'].reset_index()
```

```python
for i in watched_movie.index:
    if watched_movie['Age_range'][i] == '10-17 years':
        watched_movie['Age_range'][i] = '10-17'
```

```python
found_out = watched_movie['Where did you hear about this movie?']
```

```python
heard_about = (','.join(found_out)).split(',')
heard_about = (';'.join(heard_about)).split(';')
```

```python
unique_modes = (pd.Series(heard_about)).unique()
len(unique_modes)
```

```python
new_df = pd.DataFrame(np.zeros(774).reshape(86, 9), columns = unique_modes)
```

```python
for i in found_out.index:
    modes = found_out[i].replace(' ;', ';')
    mode = modes.split(';')
    for x in mode:
        new_df.at[i, x] = 1
```

```python
watched = pd.concat([watched_movie, new_df], axis = 1)
watched
```
